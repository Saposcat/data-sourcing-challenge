### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'
# this filter query still seems not to work for me as it will return stuff that does not have "love" in the title

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
query_url = f'{url}api-key={nyt_api_key}&fq={filter_query}&begin_date={begin_date}&end_date={end_date}&fl={field_list}'
reviews = requests.get(query_url).json()
reviews
```

```python
# Create an empty list to store the reviews
reviews_list = []

# loop through pages 0-19
for page in range(0,20):
    
    # create query with a page number
    # API results show 10 articles at a time
    query_url = f'{url}api-key={nyt_api_key}&fq={filter_query}&begin_date={begin_date}&end_date={end_date}&fl={field_list}&page={page}'
    #print(page)
    # Make a "GET" request and retrieve the JSON
    reviews = requests.get(query_url).json()
    
    # Add a twelve second interval between queries to stay within API query limits
    completion = ((page)/20) * 100
    print("Retrieval in process. loading:", completion,"%")
    time.sleep(12)
    
    # Try and save the reviews to the reviews_list
    try:
        # loop through the reviews["response"]["docs"] and append each review to the list
        reviews_list.append(reviews["response"]["docs"])
        # Print the page that was just retrieved
        print(f'{page} is the current page number retrieved')

        # Print the page number that had no results then break from the loop
    except:
        print("ERROR ENCOUNTERED - try increasing time between queries. The page where the error occured was:",page)
        break
print("... done")
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data

# I am not sure if here 5 means that we are looking at only the very first 5
# articles in the reviews_list, or the first 5 pages, but I think it is meant
# to be the latter here or there would be no point in doing it now rather than
# earlier.
for i in range(0,5):
    print(json.dumps(reviews_list[i],indent=4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()

# Create an empty dataframe
reviews_dataframe = pd.DataFrame()

# Loop through pages in the dataframe
for page in reviews_list:
    reviews_dataframe = pd.concat([reviews_dataframe, pd.json_normalize(page)])

# Reset the index of the DataFrame
reviews_dataframe = reviews_dataframe.reset_index(drop=True)

# Display the final DataFrame
reviews_dataframe
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early

extract = lambda st: st[st.find("\u2018")+1:st.find("\u2019 Review")]
reviews_dataframe['title'] = reviews_dataframe['headline.main'].apply(extract)
reviews_dataframe

# I am not entirely confident that this function is doing quite what it is supposed to be doing.
# What does it mean that the "end string should incluce ' Review'"?
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
reviews_dataframe['keywords'] = reviews_dataframe['keywords'].apply(extract_keywords)
reviews_dataframe

# note: If this code has been used to convert already, it will fail if trying
# to run it again. If this happens, return to the snippet where we are 
# converting to a pandas data frame.
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
```

```python
# Create an empty list to store the results
tmdb_movies_list = []

# Create a request counter to sleep the requests after a multiple
# of 50 requests
request_counter = 1
found_counter = 0
not_found_counter = 0

# Loop through the titles
# I'm a bit confused about this part. Why are we looping through the title
# list? Wouldn't it make more sense to loop through movie titles and not
# review titles?
for title in title_list:

    # Check if we need to sleep before making a request
    if request_counter % 50 == 0:
        print(f"-------Sleeping at {request_counter} requests----------")
        time.sleep(1)

    # in case nothing is found, try replacing spaces with url code:
    #title = title.replace(" ", "%20")
    
    # Add 1 to the request counter
    request_counter += 1
    
    # Perform a "GET" request for The Movie Database
    film_results = requests.get(url+title+tmdb_key_string).json()

    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.
    try:
        # Get movie id
        id = film_results["results"][0]["id"]

        # Make a request for a the full movie details
        movie_url = f"https://api.themoviedb.org/3/movie/{id}?api_key={tmdb_api_key}"
        

        # Execute "GET" request with url
        id_find = requests.get(movie_url).json()
        
        # Extract the genre names into a list
        genre_names = [d['name'] for d in id_find["genres"]]

        # Extract the spoken_languages' English name into a list
        spoken_languages = [d['english_name'] for d in id_find["spoken_languages"]]

        # Extract the production_countries' name into a list
        production_countries = [d['name'] for d in id_find["production_countries"]]

        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
        movie_dictionary = {
        'title' : title,
        'original_title' : id_find["original_title"],
        'budget' : id_find["budget"],
        'id' : id,
        'genre': genre_names,
        'language' : id_find["original_language"],
        'homepage' : id_find["homepage"],
        'overview' : id_find["overview"],
        'popularity' : id_find["popularity"],
        'runtime' : id_find["runtime"],
        'revenue' : id_find["revenue"],
        'release_date' : id_find["release_date"],
        'vote_average' : id_find["vote_average"],
        'vote_count' : id_find["vote_count"],
        'spoken_languages': spoken_languages,
        'production_countries': production_countries
        }
        tmdb_movies_list.append(movie_dictionary)
        # Print out the title that was found
        print(f"Found {title}")
        found_counter += 1
    except:
        print(f"{title} not found.")
        not_found_counter += 1
print(f"{found_counter} movies were found")
print(f"{not_found_counter} movies were not found")

```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merged_movie_review = pd.merge(movies_dataframe, reviews_dataframe, on='title')
merged_movie_review
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
columns_to_fix = ["genre","spoken_languages","production_countries"]

# Create a list of characters to remove
characters_to_remove = ["[","{","}","]","'"]

# Loop through the list of columns to fix
for column in columns_to_fix:
    # Convert the column to type 'str'
    merged_movie_review[column] = merged_movie_review[column].astype(str)
    #print(type(merged_movie_review[column]))

    # Loop through characters to remove
    for character in characters_to_remove:
        merged_movie_review[column]= merged_movie_review[column].str.replace(character,"")

# Display the fixed DataFrame
merged_movie_review
```

```python
# Drop "byline.person" column
del merged_movie_review["byline.person"]

#checking to see if it worked
merged_movie_review.columns
```

```python
# Delete duplicate rows and reset index
merged_movie_review = merged_movie_review.drop_duplicates()
merged_movie_review = merged_movie_review.reset_index()
merged_movie_review
```

```python
# Export data to CSV without the index
merged_movie_review.to_csv("merged_movie_review.csv",index = False)
```
